```python
import torch
import torchvision
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from data_util.DataTransformBuilder import build_data_transform
from data_util.DataDistributer import DataPartitioner
```

```python
stats = ((0.485, 0.456, 0.406), (0.229, 0.224, 0.225))
#interpolation_type = transforms.InterpolationMode.BILINEAR
resize = 256
centercrop_size = 224
preprocess = build_data_transform(stats, resize, centercrop_size)
```

```python
def visualize_data_distribution(stats):
    rows = [[stats[i]['y'][j] for j in range(data_partitioner.n_classes)] for i in range(data_partitioner.n_clients)]
    df = pd.DataFrame(rows, columns=[f"class {i}" for i in range(data_partitioner.n_classes)])
    plt.xlabel("Number of local training data points")
    plt.ylabel("Client Indices")
    for i in range(len(df.columns)):
        left = df.iloc[:, :i].sum(axis=1)
        plt.barh(range(len(df)), df.iloc[:, i], left=left)
    plt.show()
    
def visualize_data_distribution_focus(stats):
    color_list = plt.rcParams['axes.prop_cycle'].by_key()['color']
    rows = [[stats[i]['y'][j] for j in range(data_partitioner.n_classes)] for i in range(data_partitioner.n_clients)]
    map = np.argsort(np.array(rows)*(-1), axis=1)
    r = np.array(rows)
    for i in range(len(r)):
        r[i] = r[i,map[i]]
    df = pd.DataFrame(r, columns=[f"{i}" for i in range(data_partitioner.n_classes)])
    barlist = list()
    for i in range(len(df.columns)):
        left = df.iloc[:, :i].sum(axis=1)
        #print(left)
        barlist.append(plt.barh(range(len(df)), df.iloc[:, i], left=left, color='white'))
    #barlist[5][0].set_color('#e377c2')
    for i in range(100):
        for j in range(10):
            if r[i][j] != 0:
                barlist[j][i].set_color(color_list[map[i][j]])
    plt.xlim((490, 504))
    plt.xlabel("Number of local training data points")
    plt.ylabel("Client Indices")
    plt.show()
```

# Cifar10

```python
trainset = torchvision.datasets.CIFAR10(
    root='./Dataset', train=True, download=False, transform=preprocess
)
```

## Dirichlet Alpha=0.5

```python
n_clients=100
data_partitioner = DataPartitioner(trainset, n_clients)
data_partitioner.dirichlet_split_noniid(alpha=0.5, least_samples=32, manual_seed=448)
stats = data_partitioner.get_distirbution_stats()
visualize_data_distribution(stats)
```
